```python
#Laoding the necessary libraries
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
def visualize(filename):
    """Returns visualization outputs for the AVG, MAX, MIN results for each imported dataset."""
   
    # import the data and convert to a dataframe: 'data'
    data = pd.read_csv(filepath_or_buffer = filename)
   
    # calculate the Avg, Max, Min values
    avg_result = data.mean()
    max_result = data.max()
    min_result = data.min()
   
    # Plot the line graphs for AVG, MAX, MIN
    avg_result.plot(subplots=True, label = "Average", figsize = (9,7), color = 'yellow', linewidth = 2)
    max_result.plot(subplots=True, label = "Maximum", figsize = (9,7), color = 'blue', linewidth = 2)
    min_result.plot(subplots=True, label = "Minimum", figsize = (9,7), color = 'green', linewidth = 2)
  
    # Plot the legend, x & y labels and tick marks
    plt.title("Average, Maximum & Minimum", fontsize= 16)
    plt.legend()
    plt.xlabel('Feature Numbers', fontsize = 13, fontweight = 'bold')
    plt.ylabel('Values', fontsize = 13, fontweight = 'bold')
    plt.xticks(fontsize = 11)
    plt.yticks(fontsize = 11)
    plt.show()
```

```python
def identify_issue(data):
    "The function identifies issues in the dataset."
    ident = pd.read_csv(data,header=None); #Read the dataset
    if (ident[0].max()==0 and ident[20].max()==20):
        print("I doubt the integirty of this dataset \n \n");
    elif (ident.min(axis=0).sum()==0):
        print("Why is this happening ? \n \n");
    else:
        print("Dataset looks good \n \n");
```

```python
#Reading the datasets from the folder
dataset = sorted(glob.glob('data*.csv'))
print(dataset)
```

```python
for datasets in dataset[:4]:
    print(datasets)
    visualize(datasets)
    identify_issue(datasets)
```
